```python
# 실수의 모든 자릿수 더하기
value=input("실수를 하나 입력하세요(예:345.78)>>")
num=value.replace('.','') # 소수점 제거
sum=0
sum+=int(num[0]) # 첫 번째 자릿수 더하기
sum+=int(num[1]) # 두 번째 자릿수 더하기
sum+=int(num[2]) # 세 번째 자릿수 더하기
sum+=int(num[3]) # 네 번째 자릿수 더하기
sum+=int(num[4]) # 다섯 번째 자릿수 더하기
print("입력값:", value)
print("모든 자릿수의 합:", sum)
```

```python
# count() join()
str="Hello Python"
str.count('l') # 문자열에서 'l'의 개수
str.count('o') # 문자열에서 'o'의 개수
num='12345'
'->'.join(num) # 문자열 사이에 '-' 삽입 
print('->'.join(num)) # 문자열 사이에 '-' 삽입하여 출력
```

```python
# find() index()
# 메소드에서 ~없으면 -1 반환
# 메소드에서 index()는 ~없으면 ValueError 발생
str.find('c++')
str = '자바 C 파이썬'
str.find('자바')
str.find('파이썬')
str.find('C')
str.rfind('C') # 오른쪽에서부터 검색하여 가장 마지막에 나오는 'C'의 위치 반환

eng = 'abcdefg'
eng.find('c')
eng.rfind('c')
```

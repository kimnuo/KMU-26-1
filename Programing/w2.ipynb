```python
#논리연산자
x=10
print(x<5 and x<10)
print(x<5 or x<4)
print(not(x<5 and x<10))
```

```python
#전기 사용량 구하기
use = float(input("전기 사용량(kWh)을 입력하세요: "))
less_200 = use<=200
less_400 = 200<use<=400
greater_400 = use>400
base = 910 * less_200 + 1600 * less_400 + 7300 * greater_400
print('전기 사용량(kWh): %d, 기본요금(원): %d' % (use, base))
```
